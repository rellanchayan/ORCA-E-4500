```python
# in the batting dataframe, group the rows by playerID and yearID
# (they may have played for more than 1 team)
# calculate their total number of at-bats (AB), H
# doubles (2B), triples (3B), home runs(HR), walks (BB), games (G)
# and runs (R)
# copy this to a new dataframe


# get rid of some of the old-timey data
bat = batting[batting['yearID']> 1912]
bat = bat.groupby(['playerID', 'yearID']).sum()
cols = ['AB', 'H', '2B', '3B', 'HR', 'BB', 'G', 'R']
bat = bat[cols]
```

```python
# calculate each player's batting average: H/AB
# calculate each player's on-base-percentage: (H+BB) / (AB+BB)
# calculate the number of single (1B) a player hit:
# H - 2B - 3B - HR
# calculate a player's slugging percentage:
# (1B + 2 * 2B + 3* 3B + 4* HR) / (AB)

bat['BA'] = bat['H'] / bat['AB']
bat['BA'] = bat['BA'].fillna(0.0)

bat['OBP'] = (bat['H'] + bat['BB']) / (bat['AB'] + bat['BB'])
bat['OBP'].fillna(0.0, inplace=True)

bat['1B'] = (bat['H'] - bat['2B'] - bat['3B'] - bat['HR'])

bat['SLG'] = (bat['1B'] + 2 * bat['2B'] + 3 * bat['3B'] + 4 * bat['HR']) / bat['AB']
bat['SLG'].fillna(0.0, inplace=True)
bat['OPS'] = bat['OBP'] + bat['SLG']
```

```python
# calculate the league leaders in 1B, 2B, 3B, 4B

bat.rank()
bat.groupby('yearID').rank()
bat.groupby('yearID').idxmax()
```

```python
# join/merge the dataframe you created with df_player (on player id)
# calculate each players age in a given year (just use yearID - birthYear)
# group by age and then calculate the batting average and slugging percentage
# over all players of a given age
# at which age do players have the highest batting average and slugging percentage?

play = players.set_index('playerID')
bat_comb = bat.join(play, on='playerID')
bat_comb = bat_comb.reset_index()
bat_comb['age'] = bat_comb['yearID'] - bat_comb['birthYear']
bat_comb.set_index(['playerID', 'yearID'], inplace=True)

age_perf = bat_comb.groupby('age')[['BA', 'OBP', 'SLG', 'OPS']].mean()
```

```python
# join the salary table with the player table and the table you made
# where you calculated OBP and SLG, etc.
# calculate the total runs scored by a player of a given age
# for every year, what is average cost of a run scored for each age?

# what years does money cover?
money['yearID'].min()
# are there any duplicates?
money.duplicated().sum()

# group by player/year and calculate their total salary
mon_grp = money.groupby(['playerID', 'yearID']).sum()

# join bat_comb with mon_grp
# they have the same index, so it will perform the join on the index
mon_grp = bat_comb.join(mon_grp, how='inner')
# we could also add in the keyword on = ['playerID', 'yearID']
# to get the same result, but would perhaps be more descriptive


# age_grp is a dataframe with index age, and two columns (total) runs
# and (total) salary
age_grp = mon_grp.groupby('age')[['R', 'salary']].sum()

# apply a function to each row that divides the 2nd column (salary)
# by the first (i.e. the cost of a run in dollars)
age_grp.apply(lambda x: x[1] / x[0], axis=1)
```

```python
## advanced NaN filling

# calculate missing values by column
players.isnull().sum()

# we can fill missing values in height and weight
# by passing in a dictionary
players.fillna({'height': 70, 'weight': 170})


# alternatively, we can fill do something fancier - fill in any
# missing values for weight with the average of the non-missing weights for that year
# first we calculate mean of weight by birthYear
yr_mean = players[['birthYear', 'weight']].groupby('birthYear').mean()

# there are still some missing values - we will use fillna with the backfill method
# which means it will fill in an NaN with the next non-missing value in the dataframe
yr_mean = yr_mean.fillna(method='bfill')

# now we group by birthyear, and within each birthyear, we use fillna to map the 
# average weight for that year
players['weight'] = players.groupby('birthYear')['weight'].fillna(yr_mean['weight'])

# this function would be a generic way of doing it

def fillna_agg(df, grp_col, col, func, method):
    grp_mean = df[[grp_col, col]].groupby(grp_col).apply(func)
    grp_mean = df.fillna(method=method)
    df[col] = df.groupby(grp_col)[col].fillna(grp_mean[col])
    return df
    
fillna_agg(players, 'birthYear', 'weight', np.mean, 'bfill')
```

```python
### bucket players by weight
### calculate the career number of games played by weight

# this splits the player weights into deciles
players['weight_class'] = pd.qcut(players['weight'], q=10)

# join to players
bt_wt = bat.join(players.set_index('playerID')['weight_class'])
# drop rows with any missing values
bt_wt = bt_wt.dropna(how='any')

# do groupby and sum
bt_wt.groupby('weight_class')['G'].sum()
```
