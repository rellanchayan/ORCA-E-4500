```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rvs
import pandas as pd
import seaborn as sns
```

```python
from scipy.stats import bernoulli
from scipy.stats import binom

# bernoulli random variables

m = 200
p = 0.25

X = bernoulli.rvs(p=p, size=(m,10))
X = np.sum(X,axis=1)
Y = binom.rvs(p=p, n=10, size=(m,))

data = pd.DataFrame(data={'X':X, 'Y':Y})

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

bins = np.arange(0,11)

#sns.histplot(data=data,x='X', ax=ax1, kde=False, stat='probability', bins=bins)
#sns.histplot(data=data,x='Y', ax=ax2, kde=False, stat='probability', bins=bins)

sns.histplot(data=data,x='X', ax=ax1, kde=False, stat='probability', bins=bins, cumulative=True)
sns.histplot(data=data,x='Y', ax=ax2, kde=False, stat='probability', bins=bins, cumulative=True)


#ax1.plot(bins, binom.pmf(bins,n=10,p=p))
#ax2.plot(bins, binom.pmf(bins,n=10,p=p))

ax1.plot(bins, binom.cdf(bins,n=10,p=p))
ax2.plot(bins, binom.cdf(bins,n=10,p=p))

```

```python
from scipy.stats import norm


# bernoulli random variables

m = 100
mu = 0
sigma = 1.0

X = norm.rvs(loc=mu, scale=sigma, size=(m,))

data = pd.DataFrame(data={'X':X,})
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1,1,1)

bins = np.linspace(-5,5,50)

sns.histplot(data=data,x='X', ax=ax1, kde=False, stat='density', bins=bins)

ax1.plot(bins, norm.pdf(bins, loc=mu, scale=sigma))
mu_fit, sigma_fit = norm.fit(X)
ax1.plot(bins, norm.pdf(bins, loc=mu_fit, scale=sigma_fit))
```
